##Mengapa DISPATCH/Pemesanan Online harus di tingkatkan?


Meningkatkan layanan dispatch adalah langkah strategis untuk menjadikan taksi tradisional tetap relevan dan kompetitif di era modern. Ini tidak hanya memberikan manfaat langsung kepada pelanggan dan pengemudi tetapi juga mendukung efisiensi transportasi dan keberlanjutan kota. Beberapa alasan pentingnya meningkatkan Pemesanan dispatch adalah?

**1. Kenyamanan Teknologi yang Semakin Diterima**
- Tren Digitalisasi: Masyarakat saat ini lebih terbiasa menggunakan teknologi untuk memenuhi kebutuhan sehari-hari, termasuk transportasi. Memperkuat layanan dispatch memungkinkan taksi tradisional menyesuaikan diri dengan kebiasaan ini.
- Akses Cepat: Aplikasi berbasis dispatch memungkinkan pelanggan memesan taksi kapan saja tanpa harus berdiri di jalan mencari kendaraan.

**2. Efisiensi Transportasi**
- Pengurangan Kendaraan Kosong: Dispatch meminimalkan waktu pengemudi berkeliling tanpa penumpang, yang menghemat bahan bakar dan mengurangi kemacetan.
- Pengelolaan Armada yang Lebih Baik: Dengan sistem dispatch, pengemudi dapat diarahkan ke zona dengan permintaan tinggi, mengoptimalkan distribusi layanan.

**3. Keamanan dan Transparansi**
- Layanan dispatch sering kali dilengkapi fitur seperti pelacakan GPS, estimasi tarif, dan ulasan pelanggan, yang memberikan rasa aman dan nyaman kepada pelanggan.
- Pengemudi dan penumpang dapat memverifikasi identitas satu sama lain, mengurangi risiko kriminalitas.

**4. Peningkatan Persaingan**
- Menghadapi Ride-Hailing: Perusahaan seperti Uber dan Lyft telah mendominasi pasar karena kemudahan pemesanan online. Meningkatkan dispatch memungkinkan taksi tradisional tetap kompetitif.
- Diversifikasi Layanan: Dispatch membuka peluang untuk layanan tambahan seperti perjalanan berbagi (carpool) atau pemesanan grup.

**5. Manfaat untuk Kota**
- Pengelolaan Lalu Lintas: Dengan data yang dihasilkan dari sistem dispatch, otoritas kota dapat mempelajari pola lalu lintas dan merancang solusi transportasi yang lebih baik.
- Transportasi Ramah Lingkungan: Dengan rute yang lebih efisien dan pengurangan kendaraan kosong, dispatch berkontribusi pada penurunan emisi karbon.

##Uji Statistika

TIdak bisa dijalankan pada data yang memiliki missing values, jadi untuk sementara kita perlu hapus dulu yang missing


```python
# ambil kolom yang numerik
numerik = df.select_dtypes(include = 'number') # float dan int
```

```python
from scipy.stats import kstest

for kolom in numerik.columns:
    stat, pval = kstest(df[kolom],'norm')
    if pval < 0.05:
        print(f'Distribusi {kolom} tidak normal (Tolak H0)')
    else:
        print(f'Distribusi {kolom} normal (Terima H0)')
```

```python
from statsmodels.stats.diagnostic import lilliefors

for kolom in numerik.columns:
    stat, pval = lilliefors(df[kolom])
    if pval < 0.05:
        print(f'Distribusi {kolom} tidak normal (Tolak H0)')
    else:
        print(f'Distribusi {kolom} normal (Terima H0)')
```

```python
# Shapiro Wilk
from scipy.stats import shapiro

stat, pval = shapiro(df[kolom])
if pval < 0.05:
    print('Distribusi tidak normal (Tolak H0)')
else:
    print('Distribusi normal (Terima H0)')
```

```python
# D'Agustino Pearson
from scipy.stats import normaltest

for kolom in numerik.columns:
    stat, pval = normaltest(df[kolom])
    if pval < 0.05:
        print(f'Distribusi {kolom} tidak normal (Tolak H0)')
    else:
        print(f'Distribusi {kolom} normal (Terima H0)')
```

```python
# Gabungkan data perjalanan dengan tabel zona berdasarkan PULocationID
gabung_zona = df.merge(lok,
                            left_on='PULocationID',
                            right_on='LocationID',
                            how='left')
gabung_zona
```

```python
#MEMFILTER ZONA YANG DIBATASI
# Daftar zona yang secara eksplisit dilarang
restricted_zones = [
    "JFK Airport",
    "LaGuardia Airport"
]

# Filter data untuk Manhattan dan zona di bawah 59th Street
manhattan_zones = [
    'Alphabet City', 'Battery Park', 'Battery Park City', 'Chinatown',
    'Clinton East', 'Clinton West', 'East Chelsea', 'East Village',
    'Financial District North', 'Financial District South', 'Flatiron',
    'Garment District', 'Gramercy', 'Greenwich Village North',
    'Greenwich Village South', 'Hudson Sq', 'Kips Bay', 'Little Italy/NoLiTa',
    'Meatpacking/West Village West', 'Midtown Center', 'Midtown East',
    'Midtown North', 'Midtown South', 'Murray Hill',
    'Penn Station/Madison Sq West', 'Seaport', 'SoHo',
    'Sutton Place/Turtle Bay North', 'Times Sq/Theatre District',
    'TriBeCa/Civic Center', 'Two Bridges/Seward Park', 'UN/Turtle Bay South',
    'Union Sq', 'West Chelsea/Hudson Yards', 'West Village',
    'World Trade Center'
]

# Filter data untuk zona yang dibatasi
batas_zona = gabung_zona[
    (gabung_zona['Zone'].isin(restricted_zones)) |
    (gabung_zona['Zone'].isin(manhattan_zones))
]

# Output data hasil filter sebagai DataFrame
zona_larangan = gabung_zona.loc[batas_zona.index]
zona_larangan
```

```python
# Kelompokkan data berdasarkan trip_type di zona larangan
trip_distribution_restricted = zona_larangan.groupby('trip_type').size().reset_index(name='trip_count')

# Hitung persentase setiap tipe trip
total_trips = trip_distribution_restricted['trip_count'].sum()
trip_distribution_restricted['percentage'] = (trip_distribution_restricted['trip_count'] / total_trips) * 100

trip_distribution_restricted
```

```python
# Data zona non-larangan
zona_non_larangan = gabung_zona[~gabung_zona.index.isin(zona_larangan.index)]

# Kelompokkan data berdasarkan trip_type di zona non-larangan
trip_distribution_non_restricted = zona_non_larangan.groupby('trip_type').size().reset_index(name='trip_count')

# Hitung persentase setiap tipe trip
total_trips_non = trip_distribution_non_restricted['trip_count'].sum()
trip_distribution_non_restricted['percentage'] = (trip_distribution_non_restricted['trip_count'] / total_trips_non) * 100

trip_distribution_non_restricted
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gabungkan data untuk visualisasi
trip_comparison = pd.concat([
    trip_distribution_restricted.assign(zone_type='Zona Larangan'),
    trip_distribution_non_restricted.assign(zone_type='Zona Non Larangan')
])

plt.figure(figsize=(8, 4))
sns.barplot(data=trip_comparison, x='zone_type', y='percentage', hue='trip_type', palette='coolwarm')
plt.title('Perbandingan Dispatch vs Street Hail di Zona Larangan dan Non-Larangan', fontsize=14)
plt.xlabel('Jenis Zona', fontsize=12)
plt.ylabel('Persentase (%)', fontsize=12)
plt.legend(title='Trip Type')
plt.show()
```

dari hasil analisis yang telah di lakukan, larangan pemesanan menggunakan aplikasi di wilayah tertentu berpengaruh terhadap rendahnya pemilihan dispatch ,di buktikan dengan

Zona Larangan:

Proporsi perjalanan tipe Street Hail jauh lebih tinggi dibandingkan tipe Dispatch di zona larangan (misalnya bandara dan area Manhattan tertentu).

Hal ini menunjukkan bahwa larangan penggunaan aplikasi atau pembatasan tertentu secara langsung memengaruhi rendahnya penggunaan metode Dispatch di area tersebut.

Zona Non-Larangan:

Proporsi perjalanan tipe Street Hail juga mendominasi di zona non-larangan, namun perbedaannya tidak sebesar di zona larangan.

Dispatch masih memiliki proporsi lebih rendah meskipun tidak ada larangan langsung di zona tersebut.
